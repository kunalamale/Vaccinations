# india

```python
#save data into a separrate dataframe
k_in=k[k["iso_code"]=="IND"].copy()   # k Is DataFrame name
k_in
```

```python
k_in.isnull().sum()
```

```python
#Drop thr dates with missing value ,previously NaN that we filled with .
k_in.drop(k_in.index[k_in["total_vaccinations"]==0],inplace=True)
```

```python
#plot total vaccination as a function data
plt.figure(figsize=(19,6))  
sns.lineplot(data=k_in,x="date",y="total_vaccinations")
plt.title("Total Vaccination in india")
plt.xticks(rotation=0)
plt.show()
```

```python
#plot daily vaccination as a function data
plt.figure(figsize=(19,9))
sns.lineplot(data=k_in,x="date",y="daily_vaccinations")
plt.xticks(rotation=90)
plt.title("daily Vaccination is india");
```

```python
#Group by total vaccination given by country and sirt descending to  identify the top 10 countries.
v=k.groupby("country").max().sort_values("total_vaccinations",ascending=False)
v=v.iloc[:10]
v
#v=vacc_by_countrry.
```

```python
t=k.groupby("country").max().sort_values("total_vaccinations",ascending=False)
t=t.iloc[:10]
t
#t=total_vacc_by_country
```

```python
plt.figure(figsize=(17,7))
plt.bar(t.index,t.total_vaccinations)
plt.xticks(rotation=90)
plt.ylabel("Vaccination per 100")
plt.xlabel("Country")
plt.show()
```

```python
#sort by total vaccination deliverd deleverd by caountries and by vaccines.
n=k.groupby("vaccines").max().sort_values("total_vaccinations",ascending=False)
n.head()
```

```python
#get the top 10 vaccines by country for easy plotting
n=n.iloc[:10]
n
```

```python
#reset index to move vaccination from being to a column.
#this makes to easy for us to plot using seaborn,espectially if we want to sort by country.
n=n.reset_index()
n
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=n,x="vaccines",y="total_vaccinations",hue="country",dodge=False)
plt.xticks(rotation=90);
```
